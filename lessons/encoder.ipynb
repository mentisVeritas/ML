### ðŸ§¹ Removing Duplicated Columns

- **`embarked` and `embark_town` are duplicated columns**
  Both represent the port of embarkation. Since `embark_town` contains more complete data, **`embarked` is dropped**.

- **`pclass` and `class` are duplicated columns**
  Both describe the passenger class. The `pclass` column is more convenient for analysis and modeling because it is numeric. Therefore, **`class` is dropped**.

- **`alive` is a duplicate of the `survived` column**
  Since both columns contain the same information, **`alive` is dropped** to avoid redundancy.

```python
df.drop(['embarked', 'class', 'alive'], axis=1, inplace=True)
df
```

```python
df['embark_town'].fillna(df['embark_town'].mode()[0], inplace=True)
```
