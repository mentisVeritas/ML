```python
import numpy as np
```

```python
def matrix_multiply(A, B):
    if A.shape[1] != B.shape[0]:
        print("Error")
        return

    result = np.empty((A.shape[0], B.shape[1]))

    for r in range(A.shape[0]):
        for c in range(B.shape[1]):
            result[r, c] = np.sum(A[r, :] * B[:, c])


    print(result)

arr_1 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

arr_2 = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

matrix_multiply(arr_1, arr_2)

```
