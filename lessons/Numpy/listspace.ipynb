```python
arr3 = np.arange(100).astype(np.int8).reshape(10,10)
print(arr3)
```

```python
arr0 = np.empty(100).astype(np.int8).reshape(10,10)
np.fill_diagonal(arr0, np.diagonal(arr3))
np.fill_diagonal(np.fliplr(arr0), np.fliplr(arr3).diagonal())

print("arr3:")
print(arr3)

print("\narr0:")
print(arr0)
```

```python
main_diag = np.diag(arr3)
anti_diag = np.diag(np.fliplr(arr3))
```
