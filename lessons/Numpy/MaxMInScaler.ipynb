```python
import pandas as pd
import numpy as np
import time
import math

from scipy.fft import ifft2

df = pd.read_csv(f"job_placement.csv")
```

```python
salary = df['salary']
```

```python
salary
```

```python
def normalization(data):
    data = [i for i in data if i != 0]
    data = sorted(data)
    for i in range(len(data)):
        data[i] = (data[i] - data[0]) / (data[-1] - data[0])
    return pd.DataFrame(data)
```

```python
start = time.time()

salary = normalization(salary)

end = time.time()

print(end - start, 'seconds')

```
