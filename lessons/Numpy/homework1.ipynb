```python
import numpy as np
import pandas as pd

n = 100
np.random.seed(42)

ID = np.arange(1, n + 1)

Age = np.round(np.random.normal(22, 3, n)).astype(int)
Age = np.clip(Age, 18, 35)

Course = np.random.choice([1, 2, 3, 4], n, p=[0.3, 0.3, 0.25, 0.15])

Grade = np.round(np.clip(np.random.normal(3.8, 0.8, n), 1, 5))

Attendance = np.clip(np.random.normal(80 + (Grade - 3) * 5, 10, n), 50, 100).astype(int)

Activity = np.clip(Attendance + np.random.randn(n) * 5, 0, 100).astype(int)

# 7Ô∏è‚É£ Height ‚Äî –º–µ—Ç—Ä—ã, –Ω–æ—Ä–º–∞–ª—å–Ω–æ–µ —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ
Height = np.round(np.random.normal(1.73, 0.07, n), 2)

# 8Ô∏è‚É£ Weight ‚Äî –∫–≥, –ª–æ–≥–∏—á–Ω–æ —Å–≤—è–∑–∞–Ω —Å —Ä–æ—Å—Ç–æ–º
Weight = np.round(np.random.normal(Height * 100 - 90 + np.random.randn(n) * 5, 1), 1)

# 9Ô∏è‚É£ BMI ‚Äî –∏–Ω–¥–µ–∫—Å –º–∞—Å—Å—ã —Ç–µ–ª–∞ (—Ñ–æ—Ä–º—É–ª–∞ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–Ω–∞—è)
BMI = np.round(Weight / (Height ** 2), 1)

# üîü Grand ‚Äî —Ñ–æ—Ä–º—É–ª–∞ –∏–∑ —Ç–≤–æ–µ–π –ª–æ–≥–∏–∫–∏
# (3*Grade + 2*Attendance + Activity/3), –Ω–æ—Ä–º–∞–ª–∏–∑—É–µ–º –∫ 0‚Äì100
raw_score = (Grade * 3 + Attendance * 2 + Activity) / 3
Grand = np.round((raw_score - raw_score.min()) / (raw_score.max() - raw_score.min()) * 100, 1)

# üß© –°–æ–±–∏—Ä–∞–µ–º –≤—Å—ë –≤ DataFrame
df = pd.DataFrame({
    "ID": ID,
    "Age": Age,
    "Course": Course,
    "Grade": Grade,
    "Attendance": Attendance,
    "Activity": Activity,
    "Height": Height,
    "Weight": Weight,
    "BMI": BMI,
    "Grand": Grand
})

df
```
