```python
import requests
import pandas as pd
```

```python
page = 100
country = 'us'
app_id = '25d45046'
app_key = '7bd6e22083f1d7657b1d2a572615a15c'
key_word = 'python'

url = f"https://api.adzuna.com/v1/api/jobs/{country}/search/{page}"

param = {
    'app_id': f'{app_id}',
    'app_key': f'{app_key}',
    'what': f'{key_word}',
    "results_per_page": page,
    'content-type': 'application/json'
}

data = requests.get(url, params=param)

if data.status_code == 200:
    data = data.json().get('results')
    df = pd.json_normalize(data)
    print('OK')
else:
    print('Error')
```
