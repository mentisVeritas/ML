```python
def fetch_adzuna_jobs(keyword, country, results_per_page=50):

    url = f"https://api.adzuna.com/v1/api/jobs/{country}/search/1"
    params = {
        "app_id": app_id,
        "app_key": app_key,
        "what": keyword,
        "results_per_page": results_per_page
    }
    try:
        response = requests.get(url, params=params)
        data = response.json()
        results = data.get("results", [])
        for job in results:
            job["keyword"] = keyword
        return results
    except requests.RequestException as e:
        print(f"Ошибка запроса для {country}/{keyword}: {e}")
        return []


all_jobs = []
countries = [
    "au", "at", "be", "br", "ca", "ch", "fr", "de", "es", "in",
    "it", "mx", "nl", "nz", "pl", "sg", "za", "gb", "uk", "us"
]

keywords = [
    "data analyst",
    "software engineer",
    "marketing specialist",
    "project manager",
    "product manager",
    "graphic designer",
    "ux/ui designer",
    "data scientist",
    "financial analyst",
    "sales manager"
]

for country in countries:
    for kw in keywords:
        print(f"Запрос — страна {country}, профессия '{kw}'")
        jobs = fetch_adzuna_jobs(kw, country)
        all_jobs.extend(jobs)
        print(f"  Получено {len(jobs)} вакансий")

df = pd.json_normalize(all_jobs)
df.shape
```

```python
from geopy.geocoders import Nominatim
import pandas as pd

df = pd.read_csv('jobs.csv')

geolocator = Nominatim(user_agent="job_data_geo_fill")

def update_location(row):
    old_country, old_state, old_area, old_city = row['country'], row['state'], row['area'], row['city']

    if pd.notna(row['latitude']) and pd.notna(row['longitude']):
        print(f"Есть координаты: {row['latitude']}, {row['longitude']}")
        try:
            location = geolocator.reverse((row['latitude'], row['longitude']), timeout=10)
            if location and 'address' in location.raw:
                address = location.raw['address']
                geo_country = address.get('country')
                if geo_country and geo_country.lower() in str(row['country']).lower():
                    # Страна совпадает — обновляем остальные поля
                    new_state = address.get('state')
                    new_area = address.get('county') or address.get('region')
                    new_city = address.get('city') or address.get('town') or address.get('village')

                    row['state'] = new_state
                    row['area'] = new_area
                    row['city'] = new_city

                    print(f"Страна совпадает: {geo_country}")
                    print(f"Старое: country={old_country}, state={old_state}, area={old_area}, city={old_city}")
                    print(f"Новое: country={geo_country}, state={new_state}, area={new_area}, city={new_city}")
                else:
                    print(f"Страна не совпадает: гео={geo_country}, датафр={row['country']} — пропускаем")
        except Exception as e:
            print(f"Ошибка геокодирования: {e}")
    else:
        print("Нет координат — пропускаем")
    return row

df = df.apply(update_location, axis=1)
```
